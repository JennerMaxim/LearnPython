```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Kampalanew.csv", encoding="latin1")
```

```python
df.head()
```

```python
df.shape
```

```python
df
```

```python
df_kampala = df.replace(['/24h', '/12h', '/6h', '/3h',
                        '/1h'], ['', '', '', '', ''], regex=True)
df_kampala.head()
```

```python
df_kampala.replace('Tr/24', 0.01, inplace=True)
df_kampala.replace('Tr/12h', 0.01, inplace=True)
df_kampala.replace('Tr/6h', 0.01, inplace=True)
df_kampala.replace('Tr/3h', 0.01, inplace=True)
df_kampala.replace('Tr', 0.01, inplace=True)
```

```python
df_kampala.describe()
```

```python
# Check for missing data
missing_data = df_kampala.isnull()
missing_data.head()
```

```python
# Counting the number of missing value
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

<b>Missing columns are TC, TdC, Hr, Nh, Precmm, Prechars</b>

```python
# Change the dtype from object to float
df_kampala["Precmm"] = df_kampala["Precmm"].astype(float)
```

```python
df_kampala.dtypes
```

```python
# Precmm
ave_Precmm = df_kampala["Precmm"].mean()
print("The average of Precmm is:", ave_Precmm)
```

```python
# Use method replace to fill the missing value in Precmm
df_kampala["Precmm"].replace(np.nan, ave_Precmm, inplace=True)
```
